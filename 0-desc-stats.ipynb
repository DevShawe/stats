### Calculated summary

```python
import math

def mean(l):
    sum = 0
    for i in l:
        sum+=i
    return sum/len(l)
def median(l):
    n = len(l)
    if n%2!=0:
        return l[int(n/2)]
    else:
        return (l[int(n/2)-1] + l[int(n/2)])/2
    
def mode(l):
    print(l)
    s = set(l)
    if len(s) == len(l):
        return False
    uniquesList = list(s)
    print(uniquesList)
    freq = [l.count(i) for i in uniquesList]
    print(freq)
    maxIndex = freq.index(max(freq))
    return uniquesList[maxIndex]
def sd(l):
    mu = mean(l)
    sum = 0
    for xi in l:
        sum += (xi-mu)**2
    var = sum/len(l)
    print(var)
    return math.sqrt(var)

l=[1,2,3,4,5]
mean(l),median(l),mode(l)
```

```python
np.mean(l),np.median(l),stats.mode(l)
```
