# Imports

```python
import matplotlib.pylab as plt
import numpy as np
import math
```

## Histogram

```python
data = [10,11,11,12,12,12,14,13,14,14,14,14,13,13,15,15,15,15,15,16,16,16,16,17,17,17,18,18,19]
plt.xticks([i for i in range(len(data))])
plt.xlim(9,20)
plt.hist(data,color='skyblue',rwidth=0.5,align='left',bins=[i for i in range(len(data)+1)])
plt.show()
```

```python
data2 = [11,12,12,12,14,13,14,14,14,14,13,13,15,15,15,16,16,16,16,17,17,17,18,19]
plt.xticks([i for i in range(len(data))])
plt.xlim(9,20)
plt.hist([data,data2],rwidth=0.5,align='left',bins=[i for i in range(len(data)+1)])
plt.show()
```

```python
plt.xticks([i for i in range(len(data))])
plt.xlim(9,19)
plt.hist(data,color='skyblue',rwidth=0.5,align='left',bins=[i for i in range(len(data)+1)],cumulative=True)
plt.show()
```

```python
plt.xticks([i for i in range(len(data))])
plt.xlim(9,19)
plt.hist(data,color='skyblue',rwidth=0.5,align='left',bins=[i for i in range(len(data)+1)],cumulative=True,density=True)
plt.show()
```

## Uniform Distribution

```python
s = np.random.uniform(-1,0,1000)
s
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s,density=True)
plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
plt.show()
```
